```python
import requests
```

```python
# using my custom server
# see swagger at http://localhost:8000/docs#/
# url = "http://127.0.0.1:8000/embed"
# payload = {
#     "texts": [
#         "this is cool"
#     ]
# }

# using server by mlflow serve
# https://mlflow.org/docs/3.3.1/ml/deployment/deploy-model-locally/#deploying-inference-server
# url = "http://127.0.0.1:5001/invocations"
# payload = {
#     "inputs": [
#         "hello world"
#     ]
# }

# using server by mlflow serve from a container (host port 5002)
url = "http://localhost:5002/invocations"
payload = {
    "inputs": [
        "hello world this is a test"
    ]
}

response = requests.post(url, json=payload)

# Print response (optional)
print("Status Code:", response.status_code)
print("Response JSON:", response.json())
```
