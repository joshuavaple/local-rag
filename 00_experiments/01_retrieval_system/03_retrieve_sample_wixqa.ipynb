```python
from qdrant_client import QdrantClient
import requests
import numpy as np


QDRANT_URL = "http://localhost:6333"
# EMBEDDING_SVC_URL = "http://127.0.0.1:8000/embed" # old
EMBEDDING_SVC_URL = "http://127.0.0.1:5002/invocations"
COLLECTION_NAME = "wixqa_corpus"
qdrant_client = QdrantClient(url=QDRANT_URL)
```

```python
query_text = "Can I start accepting payments on my site while my Wix Payments account is still under verification?"
# query_payload = {"texts": [query_text]}  # old
# query_payload = {"inputs": [query_text]}
# query_vector = np.array(requests.post(EMBEDDING_SVC_URL, json=query_payload).json())
# query_vector

query_payload = {"inputs": [query_text]} # endpoint accept {"text":[list of str]}
resp = requests.post(EMBEDDING_SVC_URL, json=query_payload)
resp.raise_for_status()
query_vector = resp.json()["predictions"][0]
```

```python
hits = qdrant_client.search(
   collection_name=COLLECTION_NAME,
   query_vector=query_vector,
   limit=5  # Return 5 closest points
)
hits
```

```python
for hit in hits:
    print(hit.payload.get("contents_chunk"))
```
